```python
from IPython.html import widgets
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
sym.init_printing()

import families
import simulators
import solvers
```

```python
# number of female children of particular genotype
girls = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e, e_G, e_g = sym.var('e, e_G, e_g')

altruistic_girls = girls[0] + girls[2]
selfish_girls = girls[1] + girls[3]
```

## Wright-Bergstrom model

```python
# conditional phenotype matching probabilities (a la Wright/Bergstrom)
SGA = e + (1 - e) * altruistic_girls / (altruistic_girls + selfish_girls)
Sga = e + (1 - e) * selfish_girls / (altruistic_girls + selfish_girls)
```

```python
wright_bergstrom_params = {'c': 5.0, 'e': 0.5,
                           'PiaA': 7.0, 'PiAA': 5.0, 'Piaa': 3.0, 'PiAa': 2.0}
```

```python
# create an instance of the Family class
wright_bergstrom = families.OneMaleTwoFemales(wright_bergstrom_params,
                                              SGA=SGA,
                                              Sga=Sga)

# create an instance of the Simulator class
sim = simulators.Simulator(wright_bergstrom)
```

```python
simulators.plot_isolated_subpopulations_simulation(sim, 0.05, rtol=1e-4, **wright_bergstrom_params)
plt.show()
```

```python
# widget for varying the initial condition
eps = 1e-2
mGA0_widget = widgets.FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
e_widget = widgets.FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)

# widgets for varying the payoffs
fecundity_widget = widgets.IntSliderWidget(min=2.0, max=25, step=1, value=1.0)
exploitive_payoff_widget = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=7.0)
cooperative_payoff_widget = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=5.0)
selfish_payoff_widget = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=3.0)
naive_payoff_widget = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=2.0)
```

```python
interactive_plot = widgets.interact(simulators.plot_isolated_subpopulations_simulation,
                                    simulator=widgets.fixed(sim),
                                    mGA0=mGA0_widget,
                                    T=widgets.fixed(None),
                                    rtol=widgets.FloatTextWidget(value=1e-4),
                                    e=e_widget,
                                    c=fecundity_widget,
                                    PiaA=exploitive_payoff_widget,
                                    PiAA=cooperative_payoff_widget,
                                    Piaa=selfish_payoff_widget,
                                    PiAa=naive_payoff_widget,
                                    )
```

## Our model...

Our model has a slightly more sophisticated matching function...

```python
# Conditional phenotype matching probabilities
true_altruistic_girls = d_A * altruistic_girls
false_altruistic_girls = (1 - d_a) * selfish_girls
mistaken_for_altruistic_girls = (1 - e_G) * false_altruistic_girls
altruist_adoption_pool = true_altruistic_girls + mistaken_for_altruistic_girls

true_selfish_girls = d_a * selfish_girls
false_selfish_girls = (1 - d_A) * altruistic_girls
mistaken_for_selfish_girls = (1 - e_g) * false_selfish_girls
selfish_adoption_pool = true_selfish_girls + mistaken_for_selfish_girls

SGA = true_altruistic_girls / altruist_adoption_pool
Sga = true_selfish_girls / selfish_adoption_pool
```

```python
# define some parameters
hierarchy_params = {'c': 2.0, 'e_G': 0.5, 'e_g': 0.5, 'd_A': 0.5, 'd_a': 0.5,
                    'PiaA': 9.0, 'PiAA': 5.0, 'Piaa': 3.0, 'PiAa': 2.0}

# create an instance of our model
pugh_et_al = families.OneMaleTwoFemales(params=hierarchy_params,
                                        SGA=SGA,
                                        Sga=Sga)

# create an instance of the Simulator class
sim = simulators.Simulator(pugh_et_al)
```

### Plotting model simulations

```python
simulators.plot_isolated_subpopulations_simulation(sim, mGA0=0.5, rtol=1e-4,
                                                   d_A=0.5, d_a=0.5, e_G=0.5, e_g=0.5,
                                                   c=1.0, PiaA=7, PiAA=5, Piaa=3, PiAa=2)
plt.show()
```

Create the interactive plot...

```python
interactive_plot = widgets.interact(simulators.plot_isolated_subpopulations_simulation,
                                    simulator=widgets.fixed(sim),
                                    mGA0=mGA0_widget,
                                    T=widgets.fixed(None),
                                    rtol=widgets.FloatTextWidget(value=1e-4),
                                    e_G=screening_widget_G,
                                    e_g=screening_widget_g,
                                    d_A=signaling_widget_A,
                                    d_a=signaling_widget_a,
                                    c=fecundity_widget,
                                    PiaA=exploitive_payoff_widget,
                                    PiAA=cooperative_payoff_widget,
                                    Piaa=selfish_payoff_widget,
                                    PiAa=naive_payoff_widget,
                                    )
```
