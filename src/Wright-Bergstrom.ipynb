```python
# number of female children of particular genotype
girls = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e, e_G, e_g = sym.var('e, e_G, e_g')

altruistic_girls = girls[0] + girls[2]
selfish_girls = girls[1] + girls[3]
```

## Wright-Bergstrom assortativity

$$ S_{GA} = e + (1 - e) \left(\frac{f^A}{f^A + f^a}\right)$$

* So long as assortativity is sufficiently low, our hierarchy condition is *not* necessary for generating an equilibrium with polymorphism in the $\alpha$ gene.
* Sufficiently high assortativity and sufficiently extreme hierarchy can support polymorphism in both $\gamma$ and $\alpha$ genes (irrespective of initial condition).
* So long as $\Pi_{AA} - \Pi_{aa} < \epsilon$ (where $\epsilon$ is small), low assortativity generates a corner solution with $g$ males and $a$ females; intermediate assortativity generates polymorphism in both $\gamma$ and $\alpha$ genes; high assortativity generates corner solution with $G$ males and $A$ females. The "size" of the three regimes does depend on the initial condition.

```python
# conditional phenotype matching probabilities (a la Wright/Bergstrom)
SGA = e + (1 - e) * altruistic_girls / (altruistic_girls + selfish_girls)
Sga = e + (1 - e) * selfish_girls / (altruistic_girls + selfish_girls)
```

# Selection pressure

## Measuring natural selection pressure

### Females
We define the female natural selection pressure on the $\alpha$ gene as...

$$\ln \bigg(\frac{f^A(t)}{F^A(t)}\bigg) - \ln \bigg(\frac{f^a(t)}{F^a(t)}\bigg)$$

### Males

## Measuring sexual selection pressure
We define sexual selection pressure on the $\alpha$ gene as the difference between the population share of female adults carrying a particular allele of the $\alpha$ gene in period $t$ and the population share of female children carrying that allele in period $t-1$.

$$ \text{sexual selection pressure for A} \equiv \ln \bigg(\frac{F^A(t+1)}{f^A(t)}\bigg) - \ln \bigg(\frac{2}{f(t)}\bigg) $$
$$ \text{sexual selection pressure for a} \equiv \ln \bigg(\frac{F^a(t+1)}{f^a(t)}\bigg) - \ln \bigg(\frac{2}{f(t)}\bigg) $$

Or more simply...

$$ \text{sexual selection pressure} \equiv \ln \bigg(\frac{F^A(t+1)}{f^A(t)}\bigg) - \ln \bigg(\frac{F^a(t+1)}{f^a(t)}\bigg) $$

```python
eps, N = 1e-2, 50
initial_conditions = np.linspace(eps, 1-eps, N)
assortativities = np.linspace(eps, 1-eps, N)
results = np.empty((N, N, 2))

counter = 0
for i, assortativity in enumerate(assortativities):
    for j, initial_condition in enumerate(initial_conditions):

        # run the simulation
        tmp_params = {'PiAA': 4.0, 'PiAa': 2.0, 'PiaA': 5.0, 'Piaa': 3.95,
                      'c': 2.0, 'e': assortativity}
        wright_bergstrom_simulator.family.params = tmp_params
        wright_bergstrom_simulator.initial_condition = initial_condition
        simulation = wright_bergstrom_simulator.simulate(rtol=1e-6)
        distribution = simulators.Distribution(wright_bergstrom, simulation)
        
        # store the results
        results[i, j, 0] = distribution.share_A_female_adults.iloc[-1]
        results[i, j, 1] = distribution.number_G_male_adults.iloc[-1]
        
        counter += 1
        if counter % 100 == 0:
            print("Done with {} out of {} simulations".format(counter, N**2))
```

```python
def parameter_sweep():
    eps, N = 1e-2, 50
    initial_conditions = np.linspace(eps, 1-eps, N)
    assortativities = np.linspace(eps, 1-eps, N)
    results = np.empty((N, N, 2))

    counter = 0
    for i, assortativity in enumerate(assortativities):
        for j, initial_condition in enumerate(initial_conditions):

            # run the simulation
            tmp_params = {'PiAA': 4.0, 'PiAa': 2.0, 'PiaA': 5.0, 'Piaa': 3.0,
                          'c': 2.0, 'e': assortativity}
            wright_bergstrom_simulator.family.params = tmp_params
            wright_bergstrom_simulator.initial_condition = initial_condition
            simulation = wright_bergstrom_simulator.simulate(rtol=1e-6)
            distribution = simulators.Distribution(wright_bergstrom, simulation)

            # store the results
            results[i, j, 0] = distribution.share_A_female_adults.iloc[-1]
            results[i, j, 1] = distribution.number_G_male_adults.iloc[-1]

            counter += 1
            if counter % 100 == 0:
                print("Done with {} out of {} simulations".format(counter, N**2))
                
    return results
```

```python
eps, N = 1e-2, 50
initial_conditions = np.linspace(eps, 1-eps, N)
assortativities = np.linspace(eps, 1-eps, N)
extreme_hierarchy_results = np.empty((N, N, 2))

counter = 0
for i, assortativity in enumerate(assortativities):
    for j, initial_condition in enumerate(initial_conditions):

        # run the simulation
        tmp_params = {'PiAA': 4.0, 'PiAa': 2.0, 'PiaA': 20.0, 'Piaa': 3.0,
                      'c': 2.0, 'e': assortativity}
        wright_bergstrom_simulator.family.params = tmp_params
        wright_bergstrom_simulator.initial_condition = initial_condition
        simulation = wright_bergstrom_simulator.simulate(rtol=1e-6)
        distribution = simulators.Distribution(wright_bergstrom, simulation)
        
        # store the results
        extreme_hierarchy_results[i, j, 0] = distribution.share_A_female_adults.iloc[-1]
        extreme_hierarchy_results[i, j, 1] = distribution.number_G_male_adults.iloc[-1]
        
        counter += 1
        if counter % 100 == 0:
            print("Done with {} out of {} simulations".format(counter, N**2))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 8), sharey=True)
mappable = axes[0].imshow(extreme_hierarchy_results[:, :, 0], origin='lower',
                          extent=[0, 1, 0, 1], vmin=0, vmax=1,
                          interpolation='gaussian', cmap=mpl.cm.brg_r)
axes[1].imshow(extreme_hierarchy_results[:, :, 1], origin='lower', extent=[0, 1, 0, 1],
               vmin=0, vmax=1, interpolation='gaussian', cmap=mpl.cm.brg_r)

axes[0].set_xlabel('Initial condition, $M^{GA}(0)$', fontsize=15, family='serif')
axes[1].set_xlabel('Initial condition, $M^{GA}(0)$', fontsize=15, family='serif')
axes[0].set_ylabel('Assortativity parameter, $e$', fontsize=15, family='serif')
axes[0].grid('on')

axes[0].set_title(r'Females ($A$)', fontsize=20, family='serif')
axes[1].set_title(r'Males ($G$)', fontsize=20, family='serif')
axes[1].grid('on')

title = ('Equilibrium allele shares\n' +
         r'$\Pi^{{aA}}={PiaA},\ \Pi^{{AA}}={PiAA},\ ' +
         '\Pi^{{aa}}={Piaa},\ \Pi^{{Aa}}={PiAa}$')

params = {'PiAA': 4.0, 'PiAa': 2.0, 'PiaA': 20.0, 'Piaa': 3.0}
fig.suptitle(title.format(**params), x=0.5, y=1.05,
             family='serif', fontsize=25)

# add a color bar
fig.subplots_adjust(right=0.8)
cax = fig.add_axes([0.85, 0.1, 0.075, 0.8])
fig.colorbar(mappable, cax=cax)

plt.savefig('../images/extreme_hierarchy_sweep.jpeg', bbox_inches='tight')
plt.savefig('../images/extreme_hierarchy_sweep.pdf', bbox_inches='tight')

plt.show()
```
