```python
# number of female children of particular genotype
girls = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e, e_G, e_g = sym.var('e, e_G, e_g')

altruistic_girls = girls[0] + girls[2]
selfish_girls = girls[1] + girls[3]
```

## Wright-Bergstrom assortativity

* So long as assortativity is sufficiently low, our hierarchy condition is *not* necessary for generating an equilibrium with polymorphism in the $\alpha$ gene.
* Sufficiently high assortativity and sufficitnely extreme hierarchy can support polymorphism in both $\gamma$ and $\alpha$ genes (irrespective of initial condition).
* Corner solution with `ga` so long as $\Pi_{AA} - \Pi_{aa} < \epsilon$

$$ S_{GA} = e + (1 - e) \left(\frac{f^A}{f^A + f^a}\right)$$

```python
# conditional phenotype matching probabilities (a la Wright/Bergstrom)
SGA = e + (1 - e) * altruistic_girls / (altruistic_girls + selfish_girls)
Sga = e + (1 - e) * selfish_girls / (altruistic_girls + selfish_girls)
```

```python
def _plot_adult_female_genotypes(cls, axis, share=False):
    """Plot the timepaths for individual adult female genotypes."""
    if not share:
        cls.number_GA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$GA$')
        cls.number_Ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$Ga$')
        cls.number_gA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$gA$')
        cls.number_ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$ga$')
        title = ('Number of female adults carrying\na particular genotype')
        axis.set_title(title, fontsize=20, family='serif')

    else:
        cls.share_GA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$GA$')
        cls.share_Ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$Ga$')
        cls.share_gA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$gA$')
        cls.share_ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$ga$')
        axis.set_ylim(0, 1)
        title = ('Share of female adults carrying\na particular genotype')
        axis.set_title(title, fontsize=20, family='serif')
    
    return axis


def _plot_adult_female_alpha_alleles(cls, axis, share=False):
    """Plot the timepaths for female adult alpha alleles."""
    if not share:
        cls.number_A_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$A$')
        cls.number_a_female_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                        alpha=0.5, label='$a$')
        title = ('Number of female adults carrying\n' +
                 r'a particular allele of the $\alpha$ gene')
        axis.set_title(title, fontsize=20, family='serif')
    
    else:
        cls.share_A_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5, label='$A$')
        cls.share_a_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5, label='$a$')
        axis.set_ylim(0, 1)
        title = ('Share of female adults carrying\n' +
                 r'a particular allele of the $\alpha$ gene')
        axis.set_title(title, fontsize=20, family='serif')

    return axis


def _plot_adult_female_gamma_alleles(cls, axis, share=False):
    """Plot the timepaths for female adult gamma alleles."""
    if not share:
        cls.number_G_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$G$')
        cls.number_g_female_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                        alpha=0.5, label='$g$')
        title = ('Number of female adults carrying\n' +
                 r'a particular allele of the $\gamma$ gene')
        axis.set_title(title, fontsize=20, family='serif')
    
    else:
        cls.share_G_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5,label='$G$')
        cls.share_g_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5, label='$g$')
        axis.set_ylim(0, 1)
        title = ('Share of female adults carrying\n' +
                 r'a particular allele of the $\gamma$ gene')
        axis.set_title(title, fontsize=20, family='serif')

    return axis


def plot_adult_female_simulation(simulator, axis, mGA0=0.5, T=None, rtol=None,
                                 kind='genotypes', share=False, **params):
    
    # simulate the model
    simulator.family.params = params
    simulator.initial_condition = mGA0
    simulation = simulator.simulate(rtol, T)
    
    # compute the distributions
    dist = Distribution(simulator.family, simulation)
    
    if kind == 'genotypes':
        _plot_adult_female_genotypes(dist, axis, share)
    elif kind == 'alpha_allele':
        _plot_adult_female_alpha_alleles(dist, axis, share)
    elif kind == 'gamma_allele':
        _plot_adult_female_gamma_alleles(dist, axis, share)
    else:
        raise ValueError
    
    axis.set_xlabel('Time', fontsize=15, family='serif')
    axis.legend(loc=0, frameon=False)
    
    return axis
```

# Selection pressure

## Measuring natural selection pressure

### Females
We define the female natural selection pressure on the $\alpha$ gene as the difference between the population share of female adults carrying a particular allele of the $\alpha$ gene in period $t$ and the population share of female children carrying that allele in period $t$. For example, for the $A$ allele of the $\alpha$ gene, the change in natural selection pressure is computed as

$$ \text{natural selection pressure} \equiv \frac{\text{average # children $F^A$}}{\text{average # children $F^a$}} $$

or more simply

$$\ln \bigg(\frac{f^A(t)}{F^A(t)}\bigg) - \ln \bigg(\frac{f^a(t)}{F^a(t)}\bigg)$$

### Males

Natural selection pressure for males could be

$$\ln m^G(t) - \ln m^g(t)$$

## Measuring sexual selection pressure
We define sexual selection pressure on the $\alpha$ gene as the difference between the population share of female adults carrying a particular allele of the $\alpha$ gene in period $t$ and the population share of female children carrying that allele in period $t-1$.

$$ \text{sexual selection pressure for A} \equiv \ln \bigg(\frac{F^A(t+1)}{f^A(t)}\bigg) - \ln \bigg(\frac{2}{f(t)}\bigg) $$
$$ \text{sexual selection pressure for a} \equiv \ln \bigg(\frac{F^a(t+1)}{f^a(t)}\bigg) - \ln \bigg(\frac{2}{f(t)}\bigg) $$

Or more simply...

$$ \text{sexual selection pressure} \equiv \ln \bigg(\frac{F^A(t+1)}{f^A(t)}\bigg) - \ln \bigg(\frac{F^a(t+1)}{f^a(t)}\bigg) $$

```python
def plot_sexual_selection_pressure_gamma(simulator, mGA0, T=None, rtol=None, **params):
    
    # simulate the model
    simulator.family.params = params
    simulator.initial_condition = mGA0
    simulation = simulator.simulate(rtol, T)
    
    G_male_adults = simulation['Male Adult Genotypes'][[0, 1]]
    G_pressure = G_male_adults.sum(axis=1).diff()
    
    g_male_adults = simulation['Male Adult Genotypes'][[1, 3]]
    g_pressure = g_male_adults.sum(axis=1).diff()
    
    fig, ax = plt.subplots(1,1, figsize=(8,6))
    G_pressure.plot(axes=ax, label='$G$', marker='.', linestyle='none', alpha=0.5)
    g_pressure.plot(axes=ax, label='$g$', marker='.', linestyle='none', alpha=0.5)
    
    ax.set_xlabel('Time', family='serif', fontsize=15)
    ax.set_ylim(-1, 1)
    ax.set_title(r'Sexual selection pressure on the male $\gamma$ gene', fontsize=20, family='serif')
    ax.legend(loc=0, frameon=False)

    return [fig, ax]
    
```
