```python
# number of female children of particular genotype
girls = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e, e_G, e_g = sym.var('e, e_G, e_g')

altruistic_girls = girls[0] + girls[2]
selfish_girls = girls[1] + girls[3]
```

## Wright-Bergstrom assortativity

$$ S_{GA} = e + (1 - e) \left(\frac{f^A}{f^A + f^a}\right)$$

* So long as assortativity is sufficiently low, our hierarchy condition is *not* necessary for generating an equilibrium with polymorphism in the $\alpha$ gene.
* Sufficiently high assortativity and sufficiently extreme hierarchy can support polymorphism in both $\gamma$ and $\alpha$ genes (irrespective of initial condition).
* So long as $\Pi_{AA} - \Pi_{aa} < \epsilon$ (where $\epsilon$ is small), low assortativity generates a corner solution with $g$ males and $a$ females; intermediate assortativity generates polymorphism in both $\gamma$ and $\alpha$ genes; high assortativity generates corner solution with $G$ males and $A$ females. The "size" of the three regimes does depend on the initial condition.

* Multiple edge equilibria

    * Purely selfish

```python
# conditional phenotype matching probabilities (a la Wright/Bergstrom)
SGA = e + (1 - e) * altruistic_girls / (altruistic_girls + selfish_girls)
Sga = e + (1 - e) * selfish_girls / (altruistic_girls + selfish_girls)
```

## Plotting shares of adult females by allele

```python
def _plot_adult_female_genotypes(cls, axis, share=False):
    """Plot the timepaths for individual adult female genotypes."""
    if not share:
        cls.number_GA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$GA$')
        cls.number_Ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$Ga$')
        cls.number_gA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$gA$')
        cls.number_ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                         alpha=0.5, label='$ga$')
    else:
        cls.share_GA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$GA$')
        cls.share_Ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$Ga$')
        cls.share_gA_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$gA$')
        cls.share_ga_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$ga$')
        axis.set_ylim(0, 1)
        
    axis.set_title('Adult females', fontsize=20, family='serif')
    
    return axis


def _plot_adult_female_alpha_alleles(cls, axis, share=False):
    """Plot the timepaths for female adult alpha alleles."""
    if not share:
        cls.number_A_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$A$')
        cls.number_a_female_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                        alpha=0.5, label='$a$')
    
    else:
        cls.share_A_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5, label='$A$')
        cls.share_a_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5, label='$a$')
        axis.set_ylim(0, 1)
        
    axis.set_title(r'Adult females ($\alpha$ alleles)', fontsize=20, family='serif')

    return axis


def _plot_adult_female_gamma_alleles(cls, axis, share=False):
    """Plot the timepaths for female adult gamma alleles."""
    if not share:
        cls.number_G_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                        alpha=0.5, label='$G$')
        cls.number_g_female_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                        alpha=0.5, label='$g$')
    
    else:
        cls.share_G_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5,label='$G$')
        cls.share_g_female_adults.plot(marker='.', linestyle='none', legend=False, ax=axis,
                                       alpha=0.5, label='$g$')
        axis.set_ylim(0, 1)
    
    axis.set_title('Adult females ($\gamma$ alleles)', fontsize=20, family='serif')

    return axis


def _plot_adult_male_genotypes(cls, axis):
    """Plot the timepaths for male adult genotypes."""
    cls.number_GA_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                   alpha=0.5, label='$GA$')
    cls.number_Ga_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                   alpha=0.5, label='$Ga$')
    cls.number_gA_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                   alpha=0.5, label='$gA$')
    cls.number_ga_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                   alpha=0.5, label='$ga$')
    axis.set_title('Adult males', fontsize=20, family='serif')
    
    return axis


def _plot_adult_male_alpha_alleles(cls, axis):
    """Plot the timepaths for male adult alpha alleles."""
    cls.number_A_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                  alpha=0.5, label='$A$')
    cls.number_a_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                  alpha=0.5, label='$a$')
    axis.set_title(r'Adult males ($\alpha$ alleles)', fontsize=20, family='serif')

    return axis


def _plot_adult_male_gamma_alleles(cls, axis):
    """Plot the timepaths for male adult gamma alleles."""
    cls.number_G_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                  alpha=0.5, label='$G$')
    cls.number_g_male_adults.plot(marker='.', linestyle='none', ax=axis, legend=False,
                                  alpha=0.5, label='$g$')
    axis.set_title('Adult males ($\gamma$ alleles)', fontsize=20, family='serif')

    return axis


def plot_adult_female_simulation(cls, axis, kind='genotypes', share=False):
        
    if kind == 'genotypes':
        _plot_adult_female_genotypes(cls, axis, share)
    elif kind == 'alpha_allele':
        _plot_adult_female_alpha_alleles(cls, axis, share)
    elif kind == 'gamma_allele':
        _plot_adult_female_gamma_alleles(cls, axis, share)
    else:
        raise ValueError
    
    axis.set_xlabel('Time', fontsize=15, family='serif')
    axis.legend(loc=0, frameon=False)
    
    return axis


def plot_adult_male_simulation(cls, axis, kind='genotypes'):
        
    if kind == 'genotypes':
        _plot_adult_male_genotypes(cls, axis)
    elif kind == 'alpha_allele':
        _plot_adult_male_alpha_alleles(cls, axis)
    elif kind == 'gamma_allele':
        _plot_adult_male_gamma_alleles(cls, axis)
    else:
        raise ValueError
    
    axis.set_xlabel('Time', fontsize=15, family='serif')
    axis.legend(loc=0, frameon=False)
    
    return axis
```

```python
def plot_isolated_subpopulations_simulation(simulator, mGA0, T=None, rtol=None,
                                            males='genotypes', females='genotypes',
                                            share=False, **params):
    """
    Plot a simulated trajectory given an initial condition consistent with the
    isolated sub-populations assumption.

    Parameters
    ----------
    simulator : simulators.Simulator
    mGA0 : float
    T : int (default=None)
        The number of time steps to simulate.
    rtol : float (default=None)
        Simulate the model until the relative difference between timesteps
        is sufficiently small.
    males : str
    females : str
    share : boolean
    params : dict

    Returns
    -------
    A list containing...

    """
    # simulate the model
    simulator.family.params = params
    simulator.initial_condition = mGA0
    simulation = simulator.simulate(rtol, T)
    
    # compute distributions
    distribution = simulators.Distribution(simulator.family, simulation)

    fig, axes = plt.subplots(1, 2, figsize=(12, 8), sharey=True)
    plot_adult_female_simulation(distribution, axes[0], females, share)
    plot_adult_male_simulation(distribution, axes[1], males)
    
    # figure title
    if not share:
        fig_title = ('Numbers when\n$m^{{GA}}(0)={0}$, $e={e}$, ' +
                     r'$\Pi^{{aA}}={PiaA},\ \Pi^{{AA}}={PiAA},\ \Pi^{{aa}}={Piaa},\ \Pi^{{Aa}}={PiAa}$')
    else:
        fig_title = ('Population shares when\n$m^{{GA}}(0)={0}$, $e={e}$, ' +
                     r'$\Pi^{{aA}}={PiaA},\ \Pi^{{AA}}={PiAA},\ \Pi^{{aa}}={Piaa},\ \Pi^{{Aa}}={PiAa}$')

    fig.suptitle(fig_title.format(mGA0, **params), x=0.5, y=1.05, fontsize=25, family='serif')

    return [fig, axes]
```

# Selection pressure

## Measuring natural selection pressure

### Females
We define the female natural selection pressure on the $\alpha$ gene as...

$$\ln \bigg(\frac{f^A(t)}{F^A(t)}\bigg) - \ln \bigg(\frac{f^a(t)}{F^a(t)}\bigg)$$

### Males

## Measuring sexual selection pressure
We define sexual selection pressure on the $\alpha$ gene as the difference between the population share of female adults carrying a particular allele of the $\alpha$ gene in period $t$ and the population share of female children carrying that allele in period $t-1$.

$$ \text{sexual selection pressure for A} \equiv \ln \bigg(\frac{F^A(t+1)}{f^A(t)}\bigg) - \ln \bigg(\frac{2}{f(t)}\bigg) $$
$$ \text{sexual selection pressure for a} \equiv \ln \bigg(\frac{F^a(t+1)}{f^a(t)}\bigg) - \ln \bigg(\frac{2}{f(t)}\bigg) $$

Or more simply...

$$ \text{sexual selection pressure} \equiv \ln \bigg(\frac{F^A(t+1)}{f^A(t)}\bigg) - \ln \bigg(\frac{F^a(t+1)}{f^a(t)}\bigg) $$


```python
def plot_alpha_sexual_selection_pressure(cls, axis):
    """Plot female and male contributions to sexual selection pressure for the alpha gene."""
    kwargs = {'ax': axis, 'linestyle': 'none', 'marker': '.', 'alpha': 0.5}
    cls.alpha_sexual_selection_pressure_males.plot(label='Males',
                                                   color='red',
                                                   **kwargs)
    cls.alpha_sexual_selection_pressure_females.plot(label='Females',
                                                     color='blue',
                                                     **kwargs)
    cls.alpha_sexual_selection_pressure.plot(label='Total',
                                             color='purple',
                                             **kwargs)
    return axis


def plot_alpha_natural_selection_pressure(cls, axis):
    """Plot female and male contributions to natural selection pressure for the alpha gene."""
    kwargs = {'ax': axis, 'linestyle': 'none', 'marker': '.', 'alpha': 0.5}
    cls.alpha_natural_selection_pressure_males.plot(label='Males',
                                                    color='red',
                                                    **kwargs)
    cls.alpha_natural_selection_pressure_females.plot(label='Females',
                                                      color='blue',
                                                      **kwargs)
    cls.alpha_natural_selection_pressure.plot(label='Total',
                                              color='purple',
                                              **kwargs)
    return axis


def plot_gamma_sexual_selection_pressure(cls, axis):
    """Plot female and male contributions to sexual selection pressure for the gamma gene."""
    kwargs = {'ax': axis, 'linestyle': 'none', 'marker': '.', 'alpha': 0.5}
    cls.gamma_sexual_selection_pressure_males.plot(label='Males',
                                                   color='red',
                                                   **kwargs)
    cls.gamma_sexual_selection_pressure_females.plot(label='Females',
                                                     color='blue',
                                                     **kwargs)
    cls.gamma_sexual_selection_pressure.plot(label='Total',
                                             color='purple',
                                             **kwargs)
    return axis


def plot_gamma_natural_selection_pressure(cls, axis):
    """Plot female and male contributions to natural selection pressure for the gamma gene."""
    kwargs = {'ax': axis, 'linestyle': 'none', 'marker': '.', 'alpha': 0.5}
    cls.gamma_natural_selection_pressure_males.plot(label='Males',
                                                    color='red',
                                                    **kwargs)
    cls.gamma_natural_selection_pressure_females.plot(label='Females',
                                                      color='blue',
                                                      **kwargs)
    cls.gamma_natural_selection_pressure.plot(label='Total',
                                              color='purple',
                                              **kwargs)
    return axis


def plot_alpha_selection_pressure(cls, axis):
    """Plot measures of alpha natural and sexual selection pressure."""
    kwargs = {'ax': axis, 'linestyle': 'none', 'marker': '.', 'alpha': 0.5}
    cls.alpha_natural_selection_pressure.plot(label='Natural',
                                              color='red',
                                              **kwargs)
    cls.alpha_sexual_selection_pressure.plot(label='Sexual',
                                             color='blue',
                                             **kwargs)
    cls.alpha_selection_pressure.plot(label='Total',
                                      color='purple',
                                      **kwargs)

    return axis


def plot_gamma_selection_pressure(cls, axis):
    """Plot measures of gamma natural and sexual selection pressure."""
        
    kwargs = {'ax': axis, 'linestyle': 'none', 'marker': '.', 'alpha': 0.5}
    cls.gamma_natural_selection_pressure.plot(label='Natural',
                                              color='red',
                                              **kwargs)
    cls.gamma_sexual_selection_pressure.plot(label='Sexual',
                                             color='blue',
                                             **kwargs)
    cls.gamma_selection_pressure.plot(label='Total',
                                      color='purple',
                                      **kwargs)
    
    return axis


def plot_selection_pressure(simulator, mGA0, T=None, rtol=None, **params):
    """Plot measures of selection pressure on the alpha and gamma genes."""
    
    # simulate the model
    simulator.family.params = params
    simulator.initial_condition = mGA0
    sim = simulator.simulate(rtol, T)
    dist = simulators.Distribution(simulator.family, sim)
    
    fig, axes = plt.subplots(3, 2, figsize=(12, 18), sharey=True)
    
    plot_alpha_natural_selection_pressure(dist, axes[0, 0])
    plot_gamma_natural_selection_pressure(dist, axes[0, 1])
    plot_alpha_sexual_selection_pressure(dist, axes[1, 0])
    plot_gamma_sexual_selection_pressure(dist, axes[1, 1])
    plot_alpha_selection_pressure(dist, axes[2, 0])
    plot_gamma_selection_pressure(dist, axes[2, 1])
    
    # plot titles
    axes[0, 0].set_title(r'Natural selection, $\alpha$', family='serif',
                         fontsize=20)
    axes[0, 1].set_title(r'Natural selection, $\gamma$', family='serif',
                         fontsize=20)
    axes[1, 0].set_title(r'Sexual selection, $\alpha$', family='serif',
                         fontsize=20)
    axes[1, 1].set_title(r'Sexual selection, $\gamma$', family='serif',
                         fontsize=20)
    axes[2, 0].set_title(r'Combined selection, $\alpha$', family='serif',
                         fontsize=20)
    axes[2, 1].set_title(r'Combined selection, $\gamma$', family='serif',
                         fontsize=20)
    #axes[0].axhline(y=0, color='k')
    #axes[0].set_xlabel('Time', fontsize=15, family='serif')
    
    for i in range(3):
        for j in range(2):
            axes[i, j].set_xlabel('Time', fontsize=15, family='serif')
            axes[i, j].axhline(y=0, color='black')
            axes[i, j].legend(loc=0, frameon=False, prop={'family': 'serif'})
    
    
    #axes[1].axhline(y=0, color='k')
    #axes[1].set_xlabel('Time', fontsize=15, family='serif')
    #axes[1].legend(loc=0, frameon=False, prop={'family': 'serif'})
    
    title = (r'Selection pressure on $\alpha$ and $\gamma$ genes when' +
             '\n$m^{{GA}}(0)={0}$, $e={e}$, ' +
             r'$\Pi^{{aA}}={PiaA},\ \Pi^{{AA}}={PiAA},\ ' +
             '\Pi^{{aa}}={Piaa},\ \Pi^{{Aa}}={PiAa}$')
    fig.suptitle(title.format(mGA0, **params), x=0.5, family='serif', fontsize=25)
    
    return [fig, axes]
    
```

```python
def parameter_sweep():
    eps, N = 1e-2, 50
    initial_conditions = np.linspace(eps, 1-eps, N)
    assortativities = np.linspace(eps, 1-eps, N)
    results = np.empty((N, N, 2))

    counter = 0
    for i, assortativity in enumerate(assortativities):
        for j, initial_condition in enumerate(initial_conditions):

            # run the simulation
            tmp_params = {'PiAA': 4.0, 'PiAa': 2.0, 'PiaA': 5.0, 'Piaa': 3.0,
                          'c': 2.0, 'e': assortativity}
            wright_bergstrom_simulator.family.params = tmp_params
            wright_bergstrom_simulator.initial_condition = initial_condition
            simulation = wright_bergstrom_simulator.simulate(rtol=1e-6)
            distribution = simulators.Distribution(wright_bergstrom, simulation)

            # store the results
            results[i, j, 0] = distribution.share_A_female_adults.iloc[-1]
            results[i, j, 1] = distribution.number_G_male_adults.iloc[-1]

            counter += 1
            if counter % 100 == 0:
                print("Done with {} out of {} simulations".format(counter, N**2))
                
    return results
```
