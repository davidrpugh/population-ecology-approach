```python
from IPython.html import widgets
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym

import families
import simulators
import solvers
```

```python
# number of female children of particular genotype
girls = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e, e_G, e_g = sym.var('e, e_G, e_g')

altruistic_girls = girls[0] + girls[2]
selfish_girls = girls[1] + girls[3]
```

## Pugh, Schaffer, Seabright (2014) matching functions...

```python
# Conditional phenotype matching probabilities
true_altruistic_girls = d_A * altruistic_girls
false_altruistic_girls = (1 - d_a) * selfish_girls
mistaken_for_altruistic_girls = (1 - e_G) * false_altruistic_girls
altruist_adoption_pool = true_altruistic_girls + mistaken_for_altruistic_girls

true_selfish_girls = d_a * selfish_girls
false_selfish_girls = (1 - d_A) * altruistic_girls
mistaken_for_selfish_girls = (1 - e_g) * false_selfish_girls
selfish_adoption_pool = true_selfish_girls + mistaken_for_selfish_girls

SGA = true_altruistic_girls / altruist_adoption_pool
Sga = true_selfish_girls / selfish_adoption_pool
```

```python
# define some parameters
hierarchy_params = {'c': 10.0, 'e_G': 0.5, 'e_g': 0.5, 'd_A': 0.5, 'd_a': 0.5,
                    'PiaA': 9.0, 'PiAA': 5.0, 'Piaa': 3.0, 'PiAa': 2.0}

# create an instance of our model
pugh_et_al = families.OneMaleTwoFemales(params=hierarchy_params,
                                        SGA=SGA,
                                        Sga=Sga)

# create an instance of the Simulator class
pugh_et_al_simulation = simulators.Simulator(pugh_et_al)
```

### Plotting model simulations

```python
simulators.plot_isolated_subpopulations_simulation(pugh_et_al_simulation, mGA0=0.5, rtol=1e-4,
                                                   d_A=0.5, d_a=0.5, e_G=0.5, e_g=0.5,
                                                   c=10.0, PiaA=7, PiAA=5, Piaa=3, PiAa=2)
plt.show()
```

Create some widgets which we will use to vary the parameters of the model...

```python
# widget for varying the initial condition
eps = 1e-2
mGA0 = widgets.FloatSliderWidget(min=eps, max=1-eps, step=eps, value=eps)
males = widgets.DropdownWidget(values=['genotypes', 'alpha_allele', 'gamma_allele'],
                               value='genotypes')
females = widgets.DropdownWidget(values=['genotypes', 'alpha_allele', 'gamma_allele'],
                                 value='genotypes')

# widgets for varying the payoffs
fecundity = widgets.IntSliderWidget(min=2.0, max=25, step=1, value=2.0)
exploitive_payoff = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=7.0)
cooperative_payoff = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=5.0)
selfish_payoff = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=3.0)
naive_payoff = widgets.FloatSliderWidget(min=0.0, max=50.0, step=0.5, value=2.0)

# widgets for varying the signaling and screening params
male_screening_G = widgets.FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.75)
male_screening_g = widgets.FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.75)
female_signaling_A = widgets.FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.75)
female_signaling_a = widgets.FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.75)
```

Create the interactive plot...

```python
interactive_plot = widgets.interact(simulators.plot_isolated_subpopulations_simulation,
                                    simulator=widgets.fixed(pugh_et_al_simulation),
                                    mGA0=mGA0,
                                    T=widgets.fixed(None),
                                    rtol=widgets.FloatTextWidget(value=1e-8),
                                    males=males,
                                    females=females,
                                    e_G=male_screening_G,
                                    e_g=male_screening_g,
                                    d_A=female_signaling_A,
                                    d_a=female_signaling_a,
                                    c=fecundity,
                                    PiaA=exploitive_payoff,
                                    PiAA=cooperative_payoff,
                                    Piaa=selfish_payoff,
                                    PiAa=naive_payoff,
                                    )
```
