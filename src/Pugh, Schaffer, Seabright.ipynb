## Pugh, Schaffer, Seabright matching
A new matching function that might be useful because it’s so simple (and because if behaves more plausibly than Wright-Bergstrom in the limit) is exponential matching:
 
$$ S(f_A,e) = \frac{d_A f_A}{d_A f_A + (1 - e_G)(1 - d_a) f_a} $$
  
Note that when d_A=d_a=0.5 (i.e., uninformative signaling) we get

1. $S(0,e) = 0$
2. $S(1,e) = 1$
3. $S(f_A,0) = f_A$
4. $S(f_A,1) = 1$
 
So same as Wright-Bergstrom except for case 1: as A females become rare, the probability of finding one goes to zero.

```python
# number of female children of particular genotype
female_offspring = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e_G, e_g = sym.var('e_G, e_g')

altruistic_female_offspring = female_offspring[0] + female_offspring[2]
selfish_female_offspring = female_offspring[1] + female_offspring[3]
total_female_offspring = altruistic_female_offspring + selfish_female_offspring

true_altruistic_female_offspring = d_A * altruistic_female_offspring
false_altruistic_female_offspring = (1 - d_a) * selfish_female_offspring
mistaken_for_altruistic_female_offspring = (1 - e_G) * false_altruistic_female_offspring
altruist_adoption_pool = true_altruistic_female_offspring + mistaken_for_altruistic_female_offspring

true_selfish_female_offspring = d_a * selfish_female_offspring
false_selfish_female_offspring = (1 - d_A) * altruistic_female_offspring
mistaken_for_selfish_female_offspring = (1 - e_g) * false_selfish_female_offspring
selfish_adoption_pool = true_selfish_female_offspring + mistaken_for_selfish_female_offspring
```

```python
# conditional phenotype matching probabilities
SGA = true_altruistic_female_offspring / altruist_adoption_pool
Sga = true_selfish_female_offspring / selfish_adoption_pool
```

```python
params = {'c': 2.0, 'e_G': 0.5, 'e_g': 0.5, 'd_A': 0.5, 'd_a': 0.5,
          'PiaA': 5.0, 'PiAA': 4.0, 'Piaa': 3.0, 'PiAa': 2.0}
```

```python
# create an instance of the Family class
family = families.OneMaleTwoFemales(params, SGA=SGA, Sga=Sga)

# create an instance of the Simulator class
simulator = simulators.Simulator(family)
```

# Parameter sweeps

```python
def parameter_sweep(eps, N, PiAA=4.0, PiAa=2.0, PiaA=5.0, Piaa=3.0):
    
    initial_conditions = np.linspace(eps, 1-eps, N)
    assortativities = np.linspace(eps, 1-eps, N)
    results = np.empty((N, N, 2))

    counter = 0
    for i, assortativity in enumerate(assortativities):
        for j, initial_condition in enumerate(initial_conditions):

            # run the simulation
            tmp_params = {'PiAA': PiAA, 'PiAa': PiAa, 'PiaA': PiaA, 'Piaa': Piaa,
                          'c': 2.0, 'd_A': 0.5, 'd_a': 0.5, 'e_G': assortativity,
                          'e_g': assortativity}
            simulator.family.params = tmp_params
            simulator.initial_condition = initial_condition
            simulation = simulator.simulate(rtol=1e-6)
            distribution = simulators.Distribution(family, simulation)

            # store the results
            results[i, j, 0] = distribution.share_A_female_adults.iloc[-1]
            results[i, j, 1] = distribution.number_G_male_adults.iloc[-1]

            counter += 1
            if counter % 10 == 0:
                print("Done with {} out of {} simulations".format(counter, N**2))
                
    return results
```

## Linear payoffs

```python
linear_results = parameter_sweep(1e-2, 10, PiAA=4.0, PiAa=2.0, PiaA=5.0, Piaa=3.0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 8), sharey=True)
mappable = axes[0].imshow(linear_results[:, :, 0], origin='lower', extent=[0, 1, 0, 1],
               vmin=0, vmax=1, interpolation='gaussian', cmap=mpl.cm.brg_r)
axes[1].imshow(linear_results[:, :, 1], origin='lower', extent=[0, 1, 0, 1],
               vmin=0, vmax=1, interpolation='gaussian', cmap=mpl.cm.brg_r)

axes[0].set_xlabel('Initial condition, $m^{GA}(0)$', fontsize=15, family='serif')
axes[1].set_xlabel('Initial condition, $m^{GA}(0)$', fontsize=15, family='serif')
axes[0].set_ylabel('Fraction of adults mating assortatively', fontsize=15, family='serif')
axes[0].grid('on')

axes[0].set_title(r'Females ($A$)', fontsize=20, family='serif')
axes[1].set_title(r'Males ($G$)', fontsize=20, family='serif')
axes[1].grid('on')

title = ('Equilibrium allele shares\n' +
         r'$\Pi^{{aA}}={PiaA},\ \Pi^{{AA}}={PiAA},\ ' +
         '\Pi^{{aa}}={Piaa},\ \Pi^{{Aa}}={PiAa}$')
fig.suptitle(title.format(**simulator.family.params), x=0.5, y=1.05,
             family='serif', fontsize=25)

# add a color bar
fig.subplots_adjust(right=0.8)
cax = fig.add_axes([0.85, 0.1, 0.075, 0.8])
fig.colorbar(mappable, cax=cax)

plt.show()
```
