```python
from IPython.html import widgets
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
sym.init_printing()

import families
import simulators
import solvers
```

```python
# number of female children of particular genotype
girls = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e, e_G, e_g = sym.var('e, e_G, e_g')

altruistic_girls = girls[0] + girls[2]
selfish_girls = girls[1] + girls[3]
```

## Wright-Bergstrom assortativity

```python
# conditional phenotype matching probabilities (a la Wright/Bergstrom)
SGA = e + (1 - e) * altruistic_girls / (altruistic_girls + selfish_girls)
Sga = e + (1 - e) * selfish_girls / (altruistic_girls + selfish_girls)
```

```python
wright_bergstrom_params = {'c': 5.0, 'e': 0.5,
                           'PiaA': 7.0, 'PiAA': 5.0, 'Piaa': 3.0, 'PiAa': 2.0}
```

```python
# create an instance of the Family class
wright_bergstrom = families.OneMaleTwoFemales(wright_bergstrom_params,
                                              SGA=SGA,
                                              Sga=Sga)

# create an instance of the Simulator class
sim = simulators.Simulator(wright_bergstrom)
```

## Our model...

Our model has a slightly more sophisticated matching function...

```python
# Conditional phenotype matching probabilities
true_altruistic_girls = d_A * altruistic_girls
false_altruistic_girls = (1 - d_a) * selfish_girls
mistaken_for_altruistic_girls = (1 - e_G) * false_altruistic_girls
altruist_adoption_pool = true_altruistic_girls + mistaken_for_altruistic_girls

true_selfish_girls = d_a * selfish_girls
false_selfish_girls = (1 - d_A) * altruistic_girls
mistaken_for_selfish_girls = (1 - e_g) * false_selfish_girls
selfish_adoption_pool = true_selfish_girls + mistaken_for_selfish_girls

SGA = true_altruistic_girls / altruist_adoption_pool
Sga = true_selfish_girls / selfish_adoption_pool
```

```python
# define some parameters
hierarchy_params = {'c': 10.0, 'e_G': 0.5, 'e_g': 0.5, 'd_A': 0.5, 'd_a': 0.5,
                    'PiaA': 9.0, 'PiAA': 5.0, 'Piaa': 3.0, 'PiAa': 2.0}

# create an instance of our model
pugh_et_al = families.OneMaleTwoFemales(params=hierarchy_params,
                                        SGA=SGA,
                                        Sga=Sga)

# create an instance of the Simulator class
pugh_et_al_simulation = simulators.Simulator(pugh_et_al)
```

```python
pugh_et_al_simulation.initial_condition = 0.5
pugh_et_al_simulation.simulate(rtol=1e-4)[4:,:]
```

```python
def plot_isolated_subpopulations_simulation(fig, axes, simulator, mGA0, T=None, rtol=None,
                                            **params):
    """
    Plot a simulated trajectory given an initial condition consistent with the
    isolated sub-populations assumption.

    Parameters
    ----------
    simulator : simulators.Simulator
    mGA0 : float
    T : int (default=None)
        The number of time steps to simulate.
    rtol : float (default=None)
        Simulate the model until the relative difference between timesteps
        is sufficiently small.

    Returns
    -------
    A list containing...

    """
    simulator.family.params = params
    simulator.initial_condition = mGA0
    tmp_traj = simulator.simulate(rtol, T)

    # male allele trajectories
    m_GA, = axes[0].plot(tmp_traj[0], color='b', linestyle='none', marker='.',
                         markeredgecolor='b', alpha=0.05)
    m_Ga, = axes[0].plot(tmp_traj[1], color='g', linestyle='none', marker='.',
                         markeredgecolor='g', alpha=0.05)
    m_gA, = axes[0].plot(tmp_traj[2], color='r', linestyle='none', marker='.',
                         markeredgecolor='r', alpha=0.05)
    m_ga, = axes[0].plot(tmp_traj[3], color='c', linestyle='none', marker='.',
                         markeredgecolor='c', alpha=0.05)

    # female allele trajectories
    f_GA, = axes[1].plot(tmp_traj[4], color='b', linestyle='none', marker='.',
                         markeredgecolor='b', alpha=0.05, label='$GA$')
    f_Ga, = axes[1].plot(tmp_traj[5], color='g', linestyle='none', marker='.',
                         markeredgecolor='g', alpha=0.05, label='$Ga$')
    f_gA, = axes[1].plot(tmp_traj[6], color='r', linestyle='none', marker='.',
                         markeredgecolor='r', alpha=0.05, label='$gA$')
    f_ga, = axes[1].plot(tmp_traj[7], color='c', linestyle='none', marker='.',
                         markeredgecolor='c', alpha=0.05, label='$ga$')

    # specify plot options
    axes[0].set_xlabel('Time', fontsize=15, family='serif')
    axes[1].set_xlabel('Time', fontsize=15, family='serif')
    axes[0].set_ylim(0, 1)
    axes[0].set_title('Males', family='serif', fontsize=20)
    axes[1].set_title('Females', family='serif', fontsize=20)
    axes[0].grid('on')
    axes[1].grid('on')
    #axes[1].legend(loc=0, frameon=False, bbox_to_anchor=(1.25, 1.0))

    fig.suptitle('Number of males and females by genotype', fontsize=25,
                 family='serif')

    return [fig, axes]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

initial_conditions = np.linspace(0.13, 1-1e-3, 100)

for initial_condition in initial_conditions:
    plot_isolated_subpopulations_simulation(fig, axes, pugh_et_al_simulation, initial_condition,
                                            T=150, d_A=0.75, d_a=0.75, e_G=0.75, e_g=0.75,
                                            c=2.0, PiaA=7, PiAA=5, Piaa=3, PiAa=2)
    
plt.show()
```
