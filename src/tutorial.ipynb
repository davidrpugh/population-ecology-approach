```python
from IPython.html import widgets
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym

import families
import simulators
import solvers
```

```python
# number of female children of particular genotype
girls = sym.DeferredVector('f')

# Female signaling probabilities
d_A, d_a = sym.var('d_A, d_a')

# Male screening probabilities
e, e_G, e_g = sym.var('e, e_G, e_g')

altruistic_girls = girls[0] + girls[2]
selfish_girls = girls[1] + girls[3]
```

## Wright-Bergstrom assortativity

* So long as assortativity is sufficiently low, our hierarchy condition is *not* necessary for generating an equilibrium with polymorphism in the $\alpha$ gene.
* Extreme hierarchy can support polymorphism in both $\gamma$ and $\alpha$ genes.
* Corner solution with `ga` so long as $\Pi_{AA} - \Pi_{aa} < \epsilon$

$$ S_{GA} = e + (1 - e) \left(\frac{f^A}{f^A + f^a}\right)$$

```python
# conditional phenotype matching probabilities (a la Wright/Bergstrom)
SGA = e + (1 - e) * altruistic_girls / (altruistic_girls + selfish_girls)
Sga = e + (1 - e) * selfish_girls / (altruistic_girls + selfish_girls)
```

```python
wright_bergstrom_params = {'c': 5.0, 'e': 0.5,
                           'PiaA': 7.0, 'PiAA': 5.0, 'Piaa': 3.0, 'PiAa': 2.0}
```

```python
# create an instance of the Family class
wright_bergstrom = families.OneMaleTwoFemales(wright_bergstrom_params,
                                              SGA=SGA,
                                              Sga=Sga)

# create an instance of the Simulator class
wright_bergstrom_simulator = simulators.Simulator(wright_bergstrom)
```

## Plotting shares of adult females by allele

```python
def plot_female_shares(simulator, mGA0, T=None, rtol=None, **params):
    
    # simulate the model
    simulator.family.params = params
    simulator.initial_condition = mGA0
    traj = simulator.simulate(rtol, T)
    
    # compute the distributions
    df = simulator.compute_family_distributions(traj)

    # sum up the number of adult females who carry the 'A' allele
    altruist_females = (df.xs(0, level='female1_genotype') +
                        df.xs(2, level='female1_genotype') +
                        df.xs(0, level='female2_genotype') +
                        df.xs(2, level='female2_genotype'))

    # sum up the number of adult females who carry the 'a' allele
    selfish_females = (df.xs(1, level='female1_genotype') +
                       df.xs(3, level='female1_genotype') +
                       df.xs(1, level='female2_genotype') +
                       df.xs(3, level='female2_genotype'))

    # normalize to get the shares
    share_altruist_females = 0.5 * altruist_females.sum(axis=0)
    share_selfish_females = 0.5 * selfish_females.sum(axis=0)

    # make the plot
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    share_altruist_females.plot(axes=ax, marker='.', color='b',
                                linestyle='none', alpha=0.5, label='$A$')
    share_selfish_females.plot(axes=ax, marker='.', color='r',
                               linestyle='none', alpha=0.5, label='$a$')
    ax.set_xlabel('Time', fontsize=15, family='serif')
    ax.set_ylabel('Population share', fontsize=15, family='serif')
    ax.set_ylim(0, 1)
    ax.set_title('Share of female adults carrying $A$ and $a$',
                 fontsize=20, family='serif')
    ax.legend(loc=0, frameon=False, bbox_to_anchor=(1.0, 1.0))
```

## Our *slightly* more sophisticated matching functions...

```python
# Conditional phenotype matching probabilities
true_altruistic_girls = d_A * altruistic_girls
false_altruistic_girls = (1 - d_a) * selfish_girls
mistaken_for_altruistic_girls = (1 - e_G) * false_altruistic_girls
altruist_adoption_pool = true_altruistic_girls + mistaken_for_altruistic_girls

true_selfish_girls = d_a * selfish_girls
false_selfish_girls = (1 - d_A) * altruistic_girls
mistaken_for_selfish_girls = (1 - e_g) * false_selfish_girls
selfish_adoption_pool = true_selfish_girls + mistaken_for_selfish_girls

SGA = true_altruistic_girls / altruist_adoption_pool
Sga = true_selfish_girls / selfish_adoption_pool
```

```python
# define some parameters
hierarchy_params = {'c': 10.0, 'e_G': 0.5, 'e_g': 0.5, 'd_A': 0.5, 'd_a': 0.5,
                    'PiaA': 9.0, 'PiAA': 5.0, 'Piaa': 3.0, 'PiAa': 2.0}

# create an instance of our model
pugh_et_al = families.OneMaleTwoFemales(params=hierarchy_params,
                                        SGA=SGA,
                                        Sga=Sga)

# create an instance of the Simulator class
pugh_et_al_simulation = simulators.Simulator(pugh_et_al)
```
